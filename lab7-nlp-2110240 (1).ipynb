```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
docs = np.array(['Mirabai has won a silver medal in weightlifting in Tokyo olympics 2021',
                 'Sindhu has won a bronze medal in badminton in Tokyo olympics',
                 'Indian hockey team is in top four team in Tokyo olympics 2021 after 40 years'])
```

```python
bag = vectorizer.fit_transform(docs)
```

```python
print(vectorizer.vocabulary_)
```

```python
print(bag.toarray())
```

```python
from gensim.models import word2vec
```

```python
corpus = [
          'Text of the first document.',
          'Text of the second document made longer.',
          'Number three.',
          'This is number four.',
]
```

```python
tokenized_sentences = [sentence.split() for sentence in corpus]
```

```python
model = word2vec.Word2Vec(tokenized_sentences, min_count=1)
```

```python
model.wv.most_similar(positive=["document"])
```
